```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
getdir = os.getcwd()
data = pd.read_excel(getdir+'/data/Superstore.xls', index_col = 0)
data.head()
```

```python
data[['Postal Code', 'Sales', 'Discount', 'Profit']].describe()
```

```python
data['Category'].value_counts()
```

```python
data = data[data['Category'] =='Furniture']
data.head()
```

```python
data = data[['Order Date', 'Sales']].sort_values(by=['Order Date'])
data = data.groupby('Order Date').sum()
data.index
```

```python
plt.figure(figsize = (20, 10))
sns.lineplot(x = data.index, y = data['Sales'])
```

```python
# 위의 형태로는 일별 추세를 확인하기 어렵기 때문에 주, 월단위로 묶어서 확인할 수 있음
```

```python
data.plot(figsize=(20, 10))
```

```python
# 최종 데이터셋
data = data.resample('MS').sum()
```

##### fbprophet 알고리즘 사용

```python
data.plot(figsize=(20, 10))
```

```python
from fbprophet import Prophet
```

```python
df = data.reset_index()
df.columns = ['ds', 'y']

df.head()
```

```python
model = Prophet()
model.fit(df)
```

```python
future = model.make_future_dataframe(periods = 12, freq = 'MS')
pred = model.predict(future)

pred = pred[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
pred.head()
```

```python
model.plot(pred)
plt.show()
```

```python
new_df = data[:'2016']
new_df = new_df.reset_index()
new_df.columns = ['ds', 'y']

model = Prophet()
model.fit(new_df)
future = model.make_future_dataframe(periods = 12, freq = 'MS')
pred = model.predict(future)
pred = pred[['ds', 'yhat']]
pred.set_index('ds', inplace=True)
pred = pred['2017':]
pred
```

```python
ax = data.plot()
pred.plot(ax=ax, figsize = (20, 10))
```

```python
pred = model.predict(future)

model.plot_components(pred)
plt.show()
```
